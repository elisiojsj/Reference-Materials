```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mean = 100 # mean of the distribution
std = 5 # standard deviation of the distribution
size_dist = 50000
```

```python
np.random.seed(28) # set a seed
obs = np.random.normal(mean, std, size_dist)
```

```python
# plot the distribution
sns.distplot(obs)
```

```python
# sampling 100 points from the distribution
sample_100 = [np.random.choice(obs) for _ in range(100)]
```

```python
# plot distribution
sns.distplot(sample_100)
# plot mean vertical line
plt.axvline(x=np.mean(sample_100), color='black')
# describe the distribution
pd.Series(sample_100).describe()
```

```python
# sampling 400 points from the distribution
sample_400 = [np.random.choice(obs) for _ in range(400)]
```

```python
sns.distplot(sample_400)
plt.axvline(x=np.mean(sample_100), color='black')
pd.Series(sample_400).describe()
```
